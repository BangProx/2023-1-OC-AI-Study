```python
import matplotlib.pyplot as plt
x = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
y = [8, 23, 52, 37, 64]
```

```python
plt.bar(x,y,color = 'lightgray', edgecolor = 'gray')
label_font = {
    'fontsize' : 14,
    'fontweight' : 'bold',
    'color' : 'gray'
}
plt.xlabel('Salesperson',labelpad = 12, fontdict = label_font)
plt.ylabel('Revenue(1000 $)',labelpad = 12, fontdict = label_font)
for i in range(0,5):
    plt.annotate(y[i],(i,y[i]+2), ha = 'center')
    
plt.axhline(45,0,1,color = 'blue',linestyle = '--')
plt.axhline(15,0,1,color = 'red',linestyle = '--',zorder = 1)
#plt.fill_between(x,0, 15, color='red', alpha=0.25,zorder = 0)
plt.show()
```
