```python
import matplotlib.pyplot as plt
x = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
y = [8, 23, 52, 37, 64]

fig, ax = plt.subplots()
ax.bar(x, y, color = 'lightgray', edgecolor = 'gray',zorder = 2)

ax.set_xlabel('Salesperson', fontsize = 14, fontweight = 'bold', color = 'gray', labelpad=12)
ax.set_ylabel('Revenue (1000 $)', fontsize = 14, fontweight = 'bold', color = 'gray', labelpad=12)

for i, j in zip(x,y):
    ax.annotate(str(j),xy = (i,j), ha = 'center', xytext = (i,j+4))

ax.axhline(y = 45, color = 'blue', ls = '--', zorder = 1)
ax.axhline(y = 15, color = 'red', ls = '--', zorder = 1)
ax.axhspan(0,15, facecolor='red', alpha=0.2)
    
plt.show()
```
