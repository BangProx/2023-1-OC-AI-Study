```python
import matplotlib.pyplot as plt
import pandas as pd

titanic = pd.read_csv('train.csv')

x = titanic['Fare']
y = titanic['Survived']

fig, ax = plt.subplots(figsize = (5,5))
ax.scatter(x,y)
ax.set_xlabel('Fare',fontsize = 14, fontweight = 'bold', labelpad = 12)
ax.set_ylabel('Survived',fontsize = 14, fontweight = 'bold', labelpad = 12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

titanic = pd.read_csv('train.csv')

x = titanic['Age']
y = titanic['Fare']

fig, ax = plt.subplots(figsize = (5,5))
ax.scatter(x,y, color = 'green')
ax.set_xlabel('Age',fontsize = 14, fontweight = 'bold', labelpad = 12)
ax.set_ylabel('Fare',fontsize = 14, fontweight = 'bold', labelpad = 12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

titanic = pd.read_csv('train.csv')

x = titanic['Pclass']
y = titanic['Survived']

fig, ax = plt.subplots(figsize = (5,5))
ax.plot(x,y)
ax.set_xlabel('Pclass',fontsize = 14, fontweight = 'bold', labelpad = 12)
ax.set_ylabel('Survived',fontsize = 14, fontweight = 'bold', labelpad = 12)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
titanic = pd.read_csv("train.csv")

class_survival = titanic.groupby(['Pclass', 'Survived'])['PassengerId'].count().unstack()
class_survival.plot(kind='barh', stacked=True)

plt.title('Survival by Passenger Class')
plt.xlabel('Number of Passengers')
plt.ylabel('PClass')

plt.show()
```
