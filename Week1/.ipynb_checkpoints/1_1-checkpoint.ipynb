```python
import pandas as pd  #pandas import    
from random import* #모듈 선언
```

```python
#num = uniform(-1,1)   #-1~ 1 사이의 난수
randlist = [uniform(-2,2)]
for i in range(39):
    randlist.append(uniform(-1,1))
    
rand_df = pd.DataFrame({'A':randlist[0:10],
                        'B':randlist[10:20],
                        'C':randlist[20:30],
                        'D':randlist[30:40]})    
```
